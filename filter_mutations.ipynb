```python
ndf = pd.DataFrame(columns = df.columns)
for index, row in df.iterrows():
    n_DNA_cov = row['n_DNA_cov_valid_bases']
    n_DNA_AF = row['n_DNA_AF']
    t_DNA_cov = row['t_DNA_cov_valid_bases']
    t_DNA_AF = row['t_DNA_AF']
    t_RAN_cov = row['t_RNA_cov_valid_bases']
    t_RNA_AF = row ['t_RNA_AF']
    in_strelka = row['in_strelka']
#     print(row)
#     if in_strelka == 'in_strelka':   
#         ndf.loc[index] = row
   
    if float(n_DNA_cov) >= 10 and n_DNA_AF < 0.05:
        if t_DNA_cov >= 10 and t_DNA_AF >= 0.1:
            ndf.loc[index] = row
        elif t_RAN_cov >= 10 and t_RNA_AF >= 0.1:
            ndf.loc[index] = row  
```
